```python
letters = ["A", "B", "C"]
numbers = [1, 2, 3]

for pair in zip(letters, numbers):
    print(pair)
```

```python
for l, n in zip(letters, numbers):
    print(l, n)
```

```python
dict(zip(letters, numbers))
```

```python
pairs = [('A', 1), ('B', 2), ('C', 3)]

letters, numbers = zip(*pairs)

print(letters)
print(numbers)
```

```python
print(zip(letters, numbers))
print(list(zip(letters, numbers)))
```

```python
lst = [3, 1, 2]
lst.sort()
print(lst.sort())
print(lst)
```

```python
lst = [3, 1, 2]
new_lst = sorted(lst)
print(lst)
print(new_lst)
```

```python
d = {'blockdmask': 400, 'equal': 300, 'apple': 200, 'dish': 100, 'cook': 500}

print(d)
print('---------------')
print(d.keys())
print('---------------')
print(d.items())
print('---------------')

# value 기준
print(sorted(d.items(), key = lambda x : x[1]))
print('---------------')

# key 기준
print(sorted(d.items(), key = lambda x: x[0]))
```
