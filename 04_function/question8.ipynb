```python
def work_hours(start, end):
    return end - start

# 9시 출근, 18시 퇴근
worked = work_hours(9, 18)
print(f"오늘 근무한 시간은 {worked}시간입니다.")
```

```python
import random

random.seed(42)

def recommend_menu(*menus):
    return random.choice(menus)

# 메뉴
selected_menu = recommend_menu("돈까스", "김치찌개", "제육볶음",
                               "비빔밥", "라면", "샌드위치", "피자")
print(f"오늘의 점심 메뉴는 {selected_menu}입니다!")
```

```python
def work_report(tasks_done):
    if tasks_done >= 10:
        status = "우수"
    elif tasks_done >= 5:
        status = "보통"
    else:
        status = "노력 필요"
    
    return tasks_done, status

tasks, status = work_report(8)
print(f"오늘 처리한 업무는 {tasks}개입니다. 성과 평가: {status}")
```

```python
def calculate_salary(base_salary, bonus):
    total_salary = base_salary + bonus
    tax = int(total_salary * 0.1)      # 세금 10% 
    net_salary = total_salary - tax    # 실수령액
    
    return net_salary, tax

# 기본급 3,000,000원, 보너스 500,000원
base = 3000000
bonus = 500000
total = base + bonus
net_salary, tax = calculate_salary(base, bonus)

print(f"세전 월급: {total:,}원, 세금: {tax:,}원, 실수령액: {net_salary:,}원")
```
