```python
import csv

data = [
    ["회사명", "직무", "결과"],
    ["회사A", "데이터 분석가", "합격"],
    ["회사B", "소프트웨어 엔지니어", "불합격"],
    ["회사C", "제품 관리자", "진행 중"]
]

with open("applications.csv", "w", encoding='utf-8', newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)

with open("applications.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```
