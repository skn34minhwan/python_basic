```python
choice = input("서울에 갈 계획이 있나요? (y/n): ")
if choice == 'y':
    print("서울로 간다!")
```

```python
choice = input("집에 있는 가구를 다 가져갈까요? (y/n): ")
if choice == 'n':
    print("서울에서 가구를 새로 산다.")
```

```python
choice = int(input("현재 소지금은 얼마인가요? (숫자 입력): "))
if choice >= 50000:
    print("기차표를 구매했다!")
```

```python
choice = input("오늘 서울 날씨는 맑음인가요? (y/n): ")
if choice == 'y':
    print("서울 구경을 나간다.")
elif choice == 'n':
    print("집에서 쉰다.")
else:
    print("잘못된 입력입니다. y 또는 n을 입력해주세요.")
```

```python
choice = input("오늘 날씨가 어떤가요? (맑음/비/눈): ")
if choice == '맑음':
    print("한강에 가서 피크닉을 즐기는 척하며 일을 한다.")
elif choice == '비':
    print("집에서 빗소리를 ASMR 삼아 일을 한다.")
elif choice == '눈':
    print("전기장판 안에서 귤을 까먹으며 일을 한다.")
else:
    print("암막 커튼 때문에 일어나지 못했다.")
```
