```python
class Capybara:
    species = "카피바라"

    def __init__(self, name, current_job):
        self.name = name
        self.current_job = current_job

    def introduce(self):
        return f"안녕하세요, {self.species}{self.name}입니다. 현재 {self.current_job} 일을 하고 있습니다. "
    
capy1 = Capybara("노민환", "AI엔지니어")  
capy2 = Capybara("환민노", "데이터분석가")  

print(capy1.introduce())
print(capy2.introduce())

print(f"capy1의 종: {capy1.species}, capy2의 종: {capy2.species}")
```

```python
class Character:
    skills = []

    def __init__(self):
        self.skills = []
    def add_skill(self, skill):
        self.skills.append(skill)

ch1 = Character()
ch2 = Character()

ch1.add_skill("Python")
ch2.add_skill("SQL")
print(f"ch1의 skills : {ch1.skills}")
print(f"ch2의 skills : {ch2.skills}")
print(f"클래스 스킬 : {Character.skills}")
```

```python
tool = "연필"

def namespace_test():
    global tool

    local_tool = "키보드"
    tool = "펜"

    print(f"함수 내부 local_tool : {local_tool}")
    print(f"함수 내부 - global tool : {tool}")

namespace_test()
print(f"함수 외부 - global tool : {tool}")
```

```python
class Toolbox():
    tools = ["노트북", "책상"]
    def __init__(self, my_tools):
        self.tools = my_tools

    def add_tool(self, new_tools):
        self.tools.append(new_tools)

my_toolbox = Toolbox(["마우스", "모니터"])
my_toolbox.add_tool("키보드")

print(f"인스턴스 도구 : {my_toolbox.tools}")
print(f"클래스 공용 도구 : {Toolbox.tools}")
```

```python
def nested_function():
    status = "작성 중"

    def inner_function():
        status = "완료"
        print(f"내부 함수에서 상태 변경 : {status}")

    print(f"내부 함수 호출 전 상태 : {status}")
    inner_function()
    print(f"내부 함수 호출 후 상태 : {status}")

nested_function()
```
