```python
class Character:
    def __init__(self, nickname, type):
        self.nickname = nickname
        self._type = type
    
    def get_type(self):
        return self._type
    
    def set_type(self, type):
        character_types = ['전사', '법사', '치유사']

        if type in character_types:
            self._type = type
        else:
            raise ValueError("잘못된 캐릭터 타입")
my_character = Character("minhwan", "전사")
print(my_character.nickname)  

try:
    print(my_character._type)
except AttributeError as e:
    print(e)

print(my_character.get_type())  
```

```python
class Capybara:
    species = "카피바라"
    def __init__(self, name, current_job):
        self.name = name
        self.current_job = current_job

    def introduce(self):
        return f"저는 {self.name}이고, 현재 직업은 {self.current_job}입니다." 

capy1 = Capybara("카피", "데이터 분석가")
capy2 = Capybara("바라", "백엔드 개발자")

print(capy1.introduce())
print(capy2.introduce())

print(f"capy1의 종 : {capy1.species}")
print(f"capy2의 종 : {capy2.species}")
```

```python
class Character:
    skills = []
    def __init__(self):
        self.skills = []

    def add_skills(self, skill):
        self.skills.append(skill)

char1 = Character()
char2 = Character()

char1.add_skills("Python")
char1.add_skills("SQL")
char2.add_skills("AWS")

print(f"char1의 스킬 : {char1.skills}")
print(f"char2의 스킬 : {char2.skills}")
print(f"클래스 속성 skills : {Character.skills}")
```

```python
class CharacterNickname:
    nickname = "기본 카피바라"

    def __init__(self, nickname):
        self.nickname = nickname

capy = CharacterNickname("취업준비생 카피")

print(f"인스턴스의 닉네임: {capy.nickname}") 
print(f"클래스의 닉네임: {CharacterNickname.nickname}")
```

```python
tool = "연필"

def namespace_test():
    global tool
    local_tool = "키보드"

    tool = "펜"

    print(f"함수 내부 local 변수 : {local_tool}")
    print(f"함수 내부에서 수정한 global 변수 : {tool}")

print(f"함수 호출 전 전역 변수 : {tool}")
namespace_test()
print(f"함수 호출 후 전역 변수 : {tool}")
```

```python
class Toolbox:
    tools = ["노트북", "책상"]

    def __init__(self, instance_tools):
        self.tools = instance_tools
    
    def add_tool(self, new_tool):
        Toolbox.tools.append(new_tool)

my_tool = Toolbox("모니터")
your_tool = Toolbox("패드")
my_tool.add_tool("안경")

print(f"나의 도구 (인스턴스): {my_tool.tools}")     
print(f"너의 도구 (인스턴스): {your_tool.tools}")  
print(f"공용 도구 (클래스): {Toolbox.tools}")
    
```
