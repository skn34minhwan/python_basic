```python
import math 

tasks = [10, 12, 8, 15, 9]

average_task = math.fsum(tasks) / len(tasks)
print(f"평균 업무량 : {average_task}")

above_average_days = [task for task in tasks if task > average_task]
count_days = len(above_average_days)

print(f"평균 업무량보다 많이 처리한 날의 개수 : {count_days}일")
```
