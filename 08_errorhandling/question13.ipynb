```python
try:
    with open("data.txt", "r", encoding="utf-8") as file:
        content = file.read()
        print("--- data.txt 내용 ---")
        print(content)
except FileNotFoundError:
    print("파일이 존재하지 않습니다.")
```

```python
try:
    age = int(input("나이를 입력하세요: "))
    if age >= 19:
        print("성인입니다.")
    else:
        print("미성년자입니다.")
except ValueError:
    print("숫자를 입력해야 합니다.")
```

```python
try :
    n1 = int(input("숫자 입력: "))
    n2 = int(input("숫자 입력: "))

    print(int(n1 / n2))
except ZeroDivisionError as e:
    print(f"0으로 나눌 수 없습니다. (오류 내용: {e})")
except ValueError as e:
    print(f"숫자를 입력해야 합니다. (오류 내용: {e})")
```

```python
file = None
try:
    file = open("log.txt", "r", encoding="utf-8")
    print(file.read()) 

except FileNotFoundError as e:
    print(f"파일이 없습니다. (오류 내용: {e})")
finally : 
    if file:
        file.close()
    print("파일 자원 정리 완료")

```

```python
try:
    number = int(input("숫자 입력: "))
except ValueError as e:
    print(f"숫자를 입력해야 합니다. (오류 내용: {e})")
else:
    result = number / 2
    print(f"입력한 숫자를 2로 나눈 결과: {result}")
```
